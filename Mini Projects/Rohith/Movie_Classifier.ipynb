```python
import pandas as pd
import numpy  as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('Movie_classification.csv')
df.head()
```

```python
df.info()
```

```python
#missing values in Time_taken
#objects in columns 3d_available and genre
```

```python
df['Time_taken'].fillna(value=df['Time_taken'].mean(),inplace=True)
df.info()
```

```python
df.describe()
```

```python
w=np.percentile(df.Twitter_hastags,[99])[0]
w
```

```python
#df[df.Twitter_hastags>w*2]=w
```

```python
df=pd.get_dummies(df,columns=['Genre','3D_available'])
```

```python
df
```

```python
df=df.drop(['3D_available_NO'], axis=1)
```

```python
df
```

```python
#df=df.drop(['Genre','3D_available'],axis=1)
```

```python
X=df.loc[:,df.columns!='Start_Tech_Oscar']
Y=df['Start_Tech_Oscar']
```

```python
from sklearn.model_selection import train_test_split as tts
X_train,X_test,Y_train,Y_test=tts(X,Y,test_size=0.2,random_state=42)
```

```python
from sklearn import tree
Decision=tree.DecisionTreeClassifier(max_depth=4)
Decision.fit(X_train,Y_train)
#min_samples_leaf,min_samples_split
```

```python
y_test_pred=Decision.predict(X_test)
```

```python
from sklearn.metrics import r2_score
r2_score(Y_test,y_test_pred)
```

```python
from sklearn.tree import export_graphviz
import graphviz

dot_data = export_graphviz(Decision, out_file=None, filled=True,feature_names=X_train.columns)
graph = graphviz.Source(dot_data)
graph
```

```python
graph.render("decision_tree")
graph.view("decision_tree")
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(Y_test,y_test_pred)
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(Y_test,y_test_pred)
```
