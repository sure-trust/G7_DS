```python
import pandas as pd
import numpy  as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('Movie_regression.csv',header=0)
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
#missing vales in time taken
#objects in columns 3d available and genre
```

```python
df['Time_taken'].fillna(value=df['Time_taken'].mean(),inplace=True)
#replaced missing values with avg values
```

```python
sns.jointplot(data=df,x='Twitter_hastags',y='Lead_ Actor_Rating')
```

```python
w=np.percentile(df.Twitter_hastags,[99])[0]
w
```

```python
df[df.Twitter_hastags>w*2]=w
```

```python
df=pd.get_dummies(df,columns=['Genre','3D_available'])
```

```python
df
```

```python
#df.drop['Genre_307.9904','Genre_Comedy','Genre_Drama','3D_available_307.9904','3D_available_NO',axis=1]
df=df.drop(['Genre_307.9904','3D_available_307.9904','3D_available_NO'], axis=1)
```

```python
df
```

```python
X=df.loc[:,df.columns!='Collection']
Y=df['Collection']
```

```python
from sklearn.model_selection import train_test_split as tts
```

```python
X_train,X_test,Y_train,Y_test=tts(X,Y,test_size=0.2,random_state=42)
```

```python
from sklearn import tree
Decision=tree.DecisionTreeRegressor(max_depth=4)
Decision.fit(X_train,Y_train)
#min_samples_leaf,min_samples_split
```

```python
y_test_pred=Decision.predict(X_test)
```

```python
from sklearn.metrics import r2_score
r2_score(Y_test,y_test_pred)
```
