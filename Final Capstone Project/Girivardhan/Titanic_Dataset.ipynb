### **Age Column**

```python
fig = px.box(df, x='Pclass', y='Age', color='Pclass',title='Age Distribution by Pclass',)
fig.show()
```

We can see the wealthier passengers in the higher classes tend to be older, which makes sense. We'll use these average age values to impute based on Pclass for Age.

```python
def impute_age(cols):
  Age=cols[0]
  Pclass=cols[1]

  if(pd.isnull(Age)):
    if(Pclass==1):
      return 37
    elif(Pclass==2):
      return 29
    else:
      return 24
  else:
    return Age

```

```python
df['Age'] = df[['Age','Pclass']].apply(impute_age,axis=1)
```

### **Fare Column**

```python
df['Fare'].fillna(df['Fare'].mean(), inplace=True)
```

## **Train Test Split**

```python
X=df.loc[:,df.columns!="Survived"]
Y=df["Survived"]
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.tree import DecisionTreeClassifier
```

```python
models = [
    ('Logistic Regression', LogisticRegression(max_iter=1000)),
    ('Random Forest Classifier', RandomForestClassifier(n_estimators=100, random_state=42)),
    ('Gradient Boosting Classifier', GradientBoostingClassifier(n_estimators=100, random_state=42)),
    ('Decision Tree Classifier', DecisionTreeClassifier(random_state=42,max_depth=7,min_samples_leaf=8))
]
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.25, random_state=42)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix

def evaluate_model(model, X_train, Y_train, X_test, Y_test):
    model_name, clf = model
    clf.fit(X_train, Y_train)
    Y_pred = clf.predict(X_test)
    accuracy = accuracy_score(Y_test, Y_pred)
    cm = confusion_matrix(Y_test, Y_pred)

    print(f'{model_name} Accuracy: {accuracy}')
    print(f'{model_name} Confusion Matrix:')
    print(cm)
    print("----------------------------------")
```

```python
for model in models:
    evaluate_model(model, X_train, Y_train, X_test, Y_test)
```

## **Model Selection:**

Based on the combination of accuracy and confusion matrices, the Decision Tree Classifier stands out as the best-performing model. It achieved the highest accuracy score of 0.8386 while maintaining a good balance between true positives and false positives, making it an excellent choice for correctly classifying passengers as survivors or non-survivors.

Therefore, we strongly recommend employing the **Decision Tree Classifier** in our project to maximize predictive accuracy and performance when working with our the Titanic dataset.
