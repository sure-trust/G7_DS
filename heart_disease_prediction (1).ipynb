REMOVING OUTLIERS

```python
def remove_outlier(col):
  sorted(col)
  q1,q3=col.quantile([0.25,0.75])
  IQR=q3-q1
  lower_limit=q1-(1.5*IQR)
  upper_limit=q3+(1.5*IQR)
  return lower_limit,upper_limit
```

```python
low,high=remove_outlier(data_set['trestbps'])
low1,high1=remove_outlier(data_set['chol'])
low2,high2=remove_outlier(data_set['thalach'])
low3,high3=remove_outlier(data_set['oldpeak'])
low4,high4=remove_outlier(data_set['ca'])
```

```python
data_set['trestbps']=np.where(data_set['trestbps']>high,high,data_set['trestbps'])
data_set['chol']=np.where(data_set['chol']>high1,high1,data_set['chol'])
data_set['thalach']=np.where(data_set['thalach']>high2,high2,data_set['thalach'])
data_set['oldpeak']=np.where(data_set['oldpeak']>high3,high3,data_set['oldpeak'])
data_set['ca']=np.where(data_set['ca']>high4,high4,data_set['ca'])
```

```python
data_set['trestbps']=np.where(data_set['trestbps']<low,low,data_set['trestbps'])
data_set['chol']=np.where(data_set['chol']<low1,low1,data_set['chol'])
data_set['thalach']=np.where(data_set['thalach']<low2,low2,data_set['thalach'])
data_set['olpeak']=np.where(data_set['oldpeak']<low3,low3,data_set['oldpeak'])
data_set['ca']=np.where(data_set['trestbps']<low4,low4,data_set['ca'])
```
